```python
import numpy as np
import pandas as pd
```

# To convert a NumPy array to a Pandas series

```python
arra1=np.random.randint(1,9,5)

print(arra1)

panda_series=pd.Series(arra1);
print("\nPanda Series are following")
print(panda_series)
```

# To convert the first column of a DataFrame as a Series

```python
dataframe=pd.DataFrame({'X':[10, 20, 30],
                        'Z':[40, 50, 60]})

panda_series=dataframe.iloc[:, 0]
print(panda_series)
```

# To create the mean and standard deviation of the data of a given Series


```python
random_arr=np.random.randint(1,9,20)
print(random_arr)
arra1=pd.Series(random_arr)

print("\nMean value of given series -- ",np.mean(arra1))
print("Standard deviation of given series -- ",np.std(arra1))
```

# To sort a given Series

```python
random_arr=np.random.randint(1,9,20)
print(random_arr)
arra1=pd.Series(random_arr)
print("Sorted given series --  \n")
print(arra1.sort_values())
```
