```python
import numpy as np
```

# To add, subtract, multiply, divide arguments element-wise:

```python
arra1 = np.random.randint(11, 20, 5)
arra2 = np.random.randint(1, 10, 5)

print("ARRAY 1 -- ",arra1)
print("ARRAY 2 -- ",arra2)

add=np.add(arra1, arra2)
sub=np.subtract(arra1, arra2)
multi=np.multiply(arra1, arra2)
div = np.divide(arra1, arra2)


print("\nADDITION OF NUMPY ARRAY -- ",add)
print("\nSUBTRACTION OF NUMPY ARRAY -- ",sub)
print("\nMULTIPLICATION OF NUMPY ARRAY -- ",multi)
print("\nDIVISION OF NUMPY ARRAY -- ",div)
```

# To round elements of the array to the nearest integer:

```python
arra1 = np.random.uniform(11.0, 20.0, 5)

print(arra1)

print("ROUND NEAREST INTEGER -- ", np.round(arra1))
```

# to calculate mean across dimension, in a 2D numpy array

```python
arra1=np.array([np.random.randint(1,5,3),np.random.randint(1,5,3)])
print(arra1)

arr_mean= np.mean(arra1, axis=1)
ar_mean= np.mean(arra1, axis=0)

print("MEAN VALUE OF ARRAY, ROW WISE -- ",arr_mean)
print("MEAN VALUE OF ARRAY, COLUMN WISE -- ",ar_mean)
```

# To calculate the difference between neighboring elements, element-wise of a given array:

```python
arra1=np.random.randint(1, 10, 5)

print(arra1)
print("Difference between neighboring elements -- ", np.diff(arra1))
```
