```python
import numpy as np
```

# To find the maximum and minimum value of a given flattened array

```python
arra1= np.random.randint(1,10,(3,3));
print(arra1)

flatten_arra=arra1.flatten()

print("\n")
print("Flatten Array -- ",flatten_arra)

print("\nMaximum value from Flatten array -- ",np.max(flatten_arra))
print("Minimum value from Flatten array -- ",np.min(flatten_arra))
```

# To compute the mean, standard deviation, and variance of a given array along the second axis

```python
arra1=np.random.randint(1,7,(2,4))
print(arra1)

mean_arr=np.mean(arra1, axis=1)
print("\nMean of Array -- ",mean_arr)
mean_arr=np.std(arra1, axis=1)
print("\nStandard Deviation of Array -- ",mean_arr)
mean_arr=np.var(arra1, axis=1)
print("\nVariance of Array -- ",mean_arr)
```
