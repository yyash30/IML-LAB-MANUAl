# Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your stock price data into a Pandas DataFrame
# Replace 'your_data.csv' with the actual path to your data file
df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/ADANIPORTS.csv')

# Convert the 'Date' column to a datetime object if it's not already in the correct format
df['Date'] = pd.to_datetime(df['Date'])

# Filter the data for the desired date range
start_date = '2007-11-27'
end_date = '2008-07-04'
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

# Create a line plot of opening and closing stock prices
plt.figure(figsize=(12, 6))  # Set the figure size
plt.plot(filtered_df['Date'], filtered_df['Open'], label='Opening Price', color='blue')
plt.plot(filtered_df['Date'], filtered_df['Close'], label='Closing Price', color='green')

# Add labels and a legend
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('Stock Prices between {} and {}'.format(start_date, end_date))
plt.legend()

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
